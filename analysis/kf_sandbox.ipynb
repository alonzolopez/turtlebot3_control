```python
def bagToData(bagpath):
    bag = rosbag.Bag(bagpath, 'r')
    baginfo = yaml.load(bag._get_yaml_info())
    
    topics = [topic for topic in bag.get_type_and_topic_info()[1].keys()]
    values = [value for value in bag.get_type_and_topic_info()[1].values()]
    types = [t[0] for t in values]
    freqs = [f[3] for f in values]
    data = dict.fromkeys(topics)
    for topic in topics:
        data[topic] = {'msgs': [], 'time': []}
    
    for topic, msg, t in bag.read_messages():
        data[topic]['msgs'].append(msg)
        data[topic]['time'].append(t)
    
    return data

data = bagToData("../rosbags/sensor_data_3_2021-03-12-19-03-25.bag")
```

```python
### Process the odom data
odomx = [msg.pose.pose.position.x for msg in data['/odom']['msgs']]
odomy = [msg.pose.pose.position.y for msg in data['/odom']['msgs']]
odomt = [t.to_sec() for t in data['/odom']['time']]

### Process the /gazebo/link_states data
# get the index for each data type
base_index = data['/gazebo/link_states']['msgs'][0].name.index('turtlebot3_waffle::base_footprint')
wheel_left_index = data['/gazebo/link_states']['msgs'][0].name.index('turtlebot3_waffle::wheel_left_link')
wheel_right_index = data['/gazebo/link_states']['msgs'][0].name.index('turtlebot3_waffle::wheel_right_link')
# extract the data from the dict into a list
gazebox = [msg.pose[base_index].position.x for msg in data['/gazebo/link_states']['msgs']]
gazeboy = [msg.pose[base_index].position.y for msg in data['/gazebo/link_states']['msgs']]
gazebot = [t.to_sec() for t in data['/gazebo/link_states']['time']]
```

```python
basexfig = go.Figure()
basexfig.add_trace(go.Scatter(x=gazebot, y=gazebox, name="Gazebo"))
basexfig.add_trace(go.Scatter(x=odomt, y=odomx, name="Odom"))
basexfig.show()

baseyfig = go.Figure()
baseyfig.add_trace(go.Scatter(x=gazebot, y=gazeboy))
baseyfig.show()
```
